```python
# Filtra os dados pela coluna 'ID_UF' pelo cÃ³digo 53 que corresponde ao DF
filtro_uf = dados_aluno['ID_UF'] == 53

# Filtra os dados pela coluna 'IN_PREENCHIMENTO_PROVA' pelo cÃ³digo 1 que corresponde a um preenchimento verdadeiro da prova
filtro_preenchimento_prova = dados_aluno['IN_PREENCHIMENTO_PROVA'] == 1

# Filtra os dados pela coluna 'IN_PREENCHIMENTO_QUESTIONARIO' pelo cÃ³digo 1 que corresponde a um preenchimento 
#verdadeiro do questionÃ¡rio
filtro_preench_quest = dados_aluno['IN_PREENCHIMENTO_QUESTIONARIO'] == 1

# Aplica ambos os filtros usando o operador lÃ³gico AND (&)
dados_aluno_df = dados_aluno[filtro_uf & filtro_preenchimento_prova & filtro_preench_quest]
```
