```python
# Filtra os dados pela coluna 'ID_UF' pelo código 53 que corresponde ao DF
filtro_uf = dados_aluno['ID_UF'] == 53

# Filtra os dados pela coluna 'IN_PREENCHIMENTO_PROVA' pelo código 1 que corresponde a um preenchimento verdadeiro da prova
filtro_preenchimento_prova = dados_aluno['IN_PREENCHIMENTO_PROVA'] == 1

# Filtra os dados pela coluna 'IN_PREENCHIMENTO_QUESTIONARIO' pelo código 1 que corresponde a um preenchimento 
#verdadeiro do questionário
filtro_preench_quest = dados_aluno['IN_PREENCHIMENTO_QUESTIONARIO'] == 1

# Aplica ambos os filtros usando o operador lógico AND (&)
dados_aluno_df = dados_aluno[filtro_uf & filtro_preenchimento_prova & filtro_preench_quest]
```
