```python
# Filtra os dados pela coluna 'ID_UF' pelo código 53 que corresponde ao DF
filtro_uf = dados_aluno['ID_UF'] == 53

# Filtra os dados pela coluna 'IN_PREENCHIMENTO_LP' pelo código 1 que corresponde a um preenchimento verdadeiro da prova
filtro_preenchimento_LP = dados_aluno['IN_PREENCHIMENTO_LP'] == 1

# Filtra os dados pela coluna 'IN_PREENCHIMENTO_MT' pelo código 1 que corresponde a um preenchimento verdadeiro da prova
filtro_preenchimento_MT = dados_aluno['IN_PREENCHIMENTO_MT'] == 1

# Filtra os dados pela coluna 'IN_PREENCHIMENTO_QUESTIONARIO' pelo código 1 que corresponde a um preenchimento 
#verdadeiro do questionário
filtro_preench_quest = dados_aluno['IN_PREENCHIMENTO_QUESTIONARIO'] == 1

# Aplica ambos os filtros usando o operador lógico AND (&)
dados_aluno_df = dados_aluno[filtro_uf & filtro_preenchimento_LP & filtro_preenchimento_MT & filtro_preench_quest]
```
